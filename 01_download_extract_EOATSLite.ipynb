```python
import requests
from pathlib import Path
```

```python
DATA_DIR = Path("/home/jovyan/EasyOneArgoTSLite") 
DATA_DIR.mkdir(parents=True, exist_ok=True)
```

```python
DOWNLOAD_URL = "https://www.seanoe.org/data/00961/107233/data/123027.tar.gz"  # e.g. "https://www.seanoe.org/.../123025..."

ZIP_PATH = DATA_DIR / "EasyOneArgoTSLite.zip"

#(no full file in memory)
import math
def download_file(url, dest):
    print(f"Downloading {url} -> {dest}")
    with requests.get(url, stream=True) as r:
        r.raise_for_status()
        total = int(r.headers.get("Content-Length", 0))
        chunk_size = 1024 * 1024  # 1 MB
        n_chunks = math.ceil(total / chunk_size) if total else None

        with open(dest, "wb") as f:
            for i, chunk in enumerate(r.iter_content(chunk_size=chunk_size), start=1):
                if chunk:
                    f.write(chunk)
                if n_chunks and i % 10 == 0:
                    print(f"{i}/{n_chunks} chunks (~{i*chunk_size/1e6:.1f} MB)")
```

```python
download_file(DOWNLOAD_URL, ZIP_PATH)
print("Download finished:", ZIP_PATH, ZIP_PATH.stat().st_size, "bytes")
```

```python
import zipfile
import tarfile
from pathlib import Path

ARCHIVE_PATH = Path("/home/jovyan/EasyOneArgoTSLite/EasyOneArgoTSLite.zip")  # drop .zip for now

def extract_archive(path, out_dir):
    path = Path(path)
    print("Archive:", path, "size:", path.stat().st_size, "bytes")

    if zipfile.is_zipfile(path):
        print("Detected ZIP archive")
        with zipfile.ZipFile(path, "r") as z:
            z.extractall(out_dir)
    elif tarfile.is_tarfile(path):
        print("Detected TAR archive")
        with tarfile.open(path, "r:*") as t:
            t.extractall(out_dir)
    else:
        raise ValueError("Archive is neither ZIP nor TAR â€“ maybe HTML or corrupt?")

extract_archive(ARCHIVE_PATH, DATA_DIR)
```
